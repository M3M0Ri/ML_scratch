```
w = 100
b = 100
```

```
def compute_model_output(x, w, b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b

    return f_wb
```

```
tmp_f_wb = compute_model_output(x_train, w, b,)
```

```
tmp_f_wb = compute_model_output(x_train, w, b,)

plt.plot(x_train, tmp_f_wb)

plt.scatter(x_train, y_train)

plt.legend()
plt.show()
```
